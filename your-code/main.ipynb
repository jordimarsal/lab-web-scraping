####  20 latest earthquakes info (date, time, latitude, longitude and region name) by the EMSC as a pandas dataframe

````python
# This is the url you will scrape in this exercise
url = 'https://m.emsc.eu/earthquake/latest.php?min_mag=n/a&max_mag=n/a&date=n/a&euromed=World'
````

````python
#your code
soup = get_soup(url)
````

````python

div = soup.find_all('table',{'style':"background-color:#D9DDDE;width:100%;height:100%;"})
div2 = soup.find_all('table',{'style':"background-color:#e5ecf9;width:100%;height:100%;"})
columns = ['Date', 'Time','Magnitude','Zone']

def div_to_list(div):
    names = [span.tr.text for span in div]
    li = []
    for text in names:
        if not 'Data ' in text:
            li.append(text[:10])
            li.append(text[10:22])
            li.append(text[24:27])
            li.append(text[28:])
    return li

brute_list = div_to_list(div)
brute_list.extend(div_to_list(div2))

li = [[],[],[],[]]
for i,el in enumerate(brute_list): 
    up = False
    ele1 = re.findall('\d+-\d+-\d+',el)
    if ele1:
        if ele1[0] in el:
            li[0].append(el)
            up = True
    if 'UTC' in el:
        li[1].append(el)
        up = True
    if len(el)==3:
        li[2].append(el)
        up = True
    if not up:
        li[3].append(el)

li_tup = [(columns[0],li[0]), (columns[1],li[1]), (columns[2],li[2]),(columns[3],li[3])]

dic_eq = dict(li_tup)

df = pd.DataFrame(dic_eq)
df
````

````python
# your code
#!pip install tweepy
import tweepy
from tweepy import API
from tweepy import OAuthHandler

CONSUMER_KEY = 'XXX'
CONSUMER_SECRET = 'XXX'
ACCESS_KEY = 'XXX-XXX'
ACCESS_SECRET = 'XXX'

auth = tweepy.OAuthHandler(CONSUMER_KEY, CONSUMER_SECRET)
auth.set_access_token(ACCESS_KEY, ACCESS_SECRET)

api = tweepy.API(auth)
````

````python


class Tweets():
    def __init__(self, username):
        self.stweets = self.query_user(username)
        self.count = len(self.stweets)
        self.text = self.rip_text()
    
    def query_user(self,username):
        query = 'from:' + username # #<hashtag>
        max_tweets = 1000
        searched_tweets = [status for status in tweepy.Cursor(api.search,
                                                       q=query,
                                                       count=1000,
                                                       since="2019-07-01",
                                                       until="2019-07-13",
                                                       result_type="mixed"
                                                       ).items(max_tweets)]
        return searched_tweets
        
    def rip_text(self):
        text_tweets = []
        for i in range(self.count):
            text_tweets.append(self.stweets[i]._json['text'])
            return text_tweets
            
    def __str__(self):
        return searched_tweets[:3]
        
    def printl(self,n):
        for i in range(n):
            print(self.stweets[i]._json['text'])
            print(self.stweets[i]._json['created_at'])
            print("------------------------")

print_tuits = Tweets('sempresaludava')
print_tuits.printl(10)

````
